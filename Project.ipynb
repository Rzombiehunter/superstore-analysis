```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

LOADING THE DATA
We are using a sample Superstore sales dataset to explore trends in revenue, profitability, and customer segmentation.

```python
## Update the file_path variable if necessary
file_path = 'C:/Users/rzomb/OneDrive/Desktop/Data Project/Superstore.csv'
data = pd.read_csv(file_path, encoding='latin1')

## Load the dataset
df = pd.read_csv(file_path, encoding='latin1')
df.head()
```

```python
## Check for missing values in data
data.isnull().sum()

##drop rows with missing values 
data.dropna(inplace=True)

## Get basic stats of data
data.describe()
```

Business Questions to Answer:
  Which products have the highest sales and profit?
 Which regions generate the most sales and profit?
 How do sales trends vary over time?

```python
##############This shows which products are the best sellers and which generate the most profit. Sometimes, best-selling products donâ€™t have the highest profit margins 


## Group by product and sum sales and profit

top_products = data.groupby('Product Name')[['Sales', 'Profit']].sum()

## Sort by sales and profit
top_sales = top_products.sort_values(by='Sales', ascending=False).head(10)
top_profit = top_products.sort_values(by='Profit', ascending=False).head(10)

## Top 10 products by Sales
plt.figure(figsize=(12,6))
sns.barplot(x=top_sales['Sales'], y=top_sales.index, palette='Blues_d')
plt.title('Top 10 Products by Sales', fontsize=16)
plt.xlabel('Total Sales', fontsize=12)
plt.ylabel('Product', fontsize=12)
plt.grid(axis='x')
plt.tight_layout()
plt.show()

## Top 10 products by Profit
plt.figure(figsize=(12,6))
sns.barplot(x=top_profit['Profit'], y=top_profit.index, palette='Greens_d')
plt.title('Top 10 Products by Profit', fontsize=16)
plt.xlabel('Total Profit', fontsize=12)
plt.ylabel('Product', fontsize=12)
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
######################### Which regions generate the most sales and profit?
region_data = data.groupby('Region')[['Sales', 'Profit']].sum()

## Reset the index of region_data to create region_data_reset
region_data_reset = region_data.reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Sales', data=region_data_reset, color='skyblue', label='Sales')
sns.barplot(x='Region', y='Profit', data=region_data_reset, color='lightgreen', label='Profit')
plt.title('Sales and Profit by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

 1. The West Region Outperforms All Others
Sales and profit are both highest in the West, suggesting:

Strong market demand

Efficient operations (possibly better shipping/logistics)

Product-market fit with high-margin items

The West is likely a strategic stronghold for this company.

2. The East Region is Also a Strong Performer
The East ranks second in both metrics.

This region presents a stable growth opportunity, especially if marketing or delivery channels are strengthened.

3. The South is Middle-of-the-Road
Moderate sales but lower profitability suggest:

Higher operational costs

Possibly heavier discounting or returns

Suboptimal product mixes (low-margin items selling more)

4. The Central Region is Underperforming
Lowest in both sales and profit, possibly even negative profit.

Potential causes:

High shipping or return costs

Poor market penetration

Excessive discounting or promotion of unprofitable products

Needs a targeted diagnostic: logistics, customer satisfaction, or localized pricing strategies might be failing.



```python
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Group by month/year
monthly_sales = data.set_index('Order Date').resample('M').sum()

# Plot sales trend over time
plt.figure(figsize=(14,6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales['Sales'], marker='o', color='darkorange')
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

This shows whether thereâ€™s seasonality or growth in sales over time. It's great for forecasting and planning. There is a clear upward trend in sales over the time period covered.

Sales spike toward year-end, particularly in November and December, which suggests seasonal shopping trends (likely holidays).

Dips occur mid-year, especially in Juneâ€“August, possibly due to vacation season or school-related cycles.

Conclusion:
The business experiences seasonal surges in Q4 â€” plan inventory, staffing, and marketing accordingly.

Use slower summer months for process improvements, promotions, or clearing old inventory.
